```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
track1 = pd.read_excel('fitts test.xlsx',sheet_name = 'track1')
mouse1 = pd.read_excel('fitts test.xlsx',sheet_name = 'mouse1')
track2 = pd.read_excel('fitts test.xlsx',sheet_name = 'track2')
mouse2 = pd.read_excel('fitts test.xlsx',sheet_name = 'mouse2')
```

```python
track1A = track1.iloc[:,0]
track1W = track1.iloc[:,1]
track1MT = track1.iloc[:,2]
track1ID = track1.iloc[:,3]
```

```python
mouse1A = mouse1.iloc[:,0]
mouse1W = mouse1.iloc[:,1]
mouse1MT = mouse1.iloc[:,2]
mouse1ID = mouse1.iloc[:,3]
```

```python
track2A = track2.iloc[:,0]
track2W = track2.iloc[:,1]
track2MT = track2.iloc[:,2]
track2ID = track2.iloc[:,3]
```

```python
mouse2A = mouse2.iloc[:,0]
mouse2W = mouse2.iloc[:,1]
mouse2MT = mouse2.iloc[:,2]
mouse2ID = mouse2.iloc[:,3]
```

```python
sns.reset_defaults()

sns.set_style('darkgrid') 
plt.rc('axes', titlesize=18)     # fontsize of the axes title
plt.rc('axes', labelsize=14)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=13)    # fontsize of the tick labels
plt.rc('ytick', labelsize=13)    # fontsize of the tick labels
plt.rc('legend', fontsize=13)    # legend fontsize
plt.rc('font', size=13)          # controls default text sizes
```

```python
fig, axs = plt.subplots(1,3, figsize=(12, 5), tight_layout=True)

title = 'First Mouse Target Parameter Dependence'
fig.suptitle('First Mouse Target Parameter Dependence')
axs[0].set_title('Amplitude Dependence')
axs[1].set_title('Width Dependence')
axs[2].set_title('ID Dependence')

sns.scatterplot(
    ax=axs[0],
    x='A',
    y='MT',
    data=mouse1

)
axs[0].set(xlabel='Amplitude (px)', ylabel='Movement Time (ms)')

sns.scatterplot(
    ax=axs[1],
    x='W',
    y='MT',
    data=mouse1
)
axs[1].set(xlabel='Width (px)', ylabel='Movement Time (ms)')


sns.scatterplot(
    ax=axs[2],
    x='ID',
    y='MT',
    data=mouse1
)

axs[2].set(xlabel='Index of Difficulty', ylabel='Movement Time (ms)')

plt.savefig(title +'.png', transparent=False)
```

```python
title = 'First Mouse Linear Regression'
slope, intercept, r_value, p_value, std_err = stats.linregress(mouse1ID,mouse1MT)

sns.regplot(x="ID", y="MT", data=mouse1, x_jitter=.05, 
           line_kws={'label':"y = {0:.1f}x + {1:.1f}".format(slope, intercept)}).legend(loc="best")




plt.savefig(title +'.png', transparent=False)
#The results show that, in agreement with Fitt's prediction, 
#a higher index of difficulty (positively correlated with target
#amplitude and negatively correlated with target width) is assocciated
#with higher movement times in the task.
```

```python

fig, axs = plt.subplots(2,2, figsize=(12, 10), tight_layout=True)

title = 'Fitts Test Scatter Plots Across Trials'
fig.suptitle('Fitts Test Scatter Plots Across Trials')
axs[0,0].set_title('First Trackpad Trial')
axs[0,1].set_title('First Mouse Trial')
axs[1,0].set_title('Second Trackpad Trial')
axs[1,1].set_title('Second Mouse Trial')

sns.scatterplot(
    ax=axs[0,0],
    x='ID',
    y='MT',
    data=track1
)
axs[0,0].set(xlabel='Index of Difficulty', ylabel='Movement Time (ms)')


sns.scatterplot(
    ax=axs[0,1],
    x='ID',
    y='MT',
    data=mouse1
)
axs[0,1].set(xlabel='Index of Difficulty', ylabel='Movement Time (ms)')

sns.scatterplot(
    ax=axs[1,0],
    x='ID',
    y='MT',
    data=track2
)
axs[1,0].set(xlabel='Index of Difficulty', ylabel='Movement Time (ms)')

sns.scatterplot(
    ax=axs[1,1],
    x='ID',
    y='MT',
    data=mouse2
)
axs[1,1].set(xlabel='Index of Difficulty', ylabel='Movement Time (ms)')


plt.savefig(title +'.png', transparent=False)

```
